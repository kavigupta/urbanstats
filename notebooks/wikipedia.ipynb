```python
def compute_text_to_add(row):
    template = (
        "As of the 2020 Census, $shortname had a [[population density]] of $awdensity/mi<sup>2</sup>"
        " and a [[population-weighted density]] of $pwdensity/mi<sup>2</sup>."
        '{{refn|Mean of density in 1mi radii circles around each person|group="note"}}'
        "<ref>{{cite web |title=$longname |url=https://urbanstats.org/article.html?longname=$urlencodedlongname"
        " |website=Urban Stats |access-date=$date}}</ref>"
    )
    # {{convert|110|/sqmi|/km2|disp=preunit|people&nbsp;|people|abbr=out}}
    mi2_to_km2 = 1.609344**2
    text = template
    text = text.replace("$shortname", row.shortname)
    text = text.replace("$longname", row.longname)
    text = text.replace("$urlencodedlongname", urllib.parse.quote_plus(row.longname))
    text = text.replace("$awdensity", f"{row.sd * mi2_to_km2:.0f}")
    text = text.replace("$pwdensity", f"{row['ad_1.609344'] * mi2_to_km2:.0f}")
    text = text.replace(
        "$date", datetime.datetime.strftime(datetime.datetime.now(), "%Y-%m-%d")
    )
    return text


def produce_instructions(row):
    print(compute_text_to_add(row))
    print()
```
